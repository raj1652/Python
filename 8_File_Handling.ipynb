```python
try:
  f = open("MyData1.txt","r")

except Exception as msg:
  print("Exception raised:",msg)

else:
  data = f.read()
  print("File Successfull Opened!!\n",data)

finally:
  print("The End --")

print("EOF")
```

```python
f = open("MyData1.txt","r")
n=1
while True:
  line = f.readline()
  if line == "":
    break
  print("Line",n,":",line)
  n+=1
```

```python
name = input("Enter name:")
gender = input("Gender:")
hobbies = input("Tell your Hobbies:")

f = open("MyData2.txt","a+")
f.write("\nName:"+name +"\n")
f.write("Gender:"+gender +"\n")
f.write("Hobbies:"+hobbies +"\n")

data = f.read() #read the whole file in 1shot
print(data)
f.close()
```

```python
# Create a copy of Mydata2 
f1 = open("MyData2.txt","r")
data = f1.read()

f2 = open("MyData3.txt","w")
f2.write(data)
f2.close()

f1.close()
```

```python
# 3rd file has alternative line from 1.txt and 2.txt

f1 = open("1.txt","r")
f2 = open("2.txt","r")

f3 = open("3.txt","w")

while True:
  p1 = f1.readline()
  p2 = f2.readline()
  if p1 == "" and p2 == "":
    break
  f3.write(p1)
  f3.write(p2)

f1.close()
f2.close()
f3.close()
```

```python
# Add a 2.txt file at end of 1.txt file
f1 = open("1.txt","a")
f2 = open("2.txt","r")
data = f2.read()
f1.write(data)

f1.close()
f2.close()
```
