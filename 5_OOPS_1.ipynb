```python
class Student:
    def __init__(self,name,rollno,pointer):
        self.rollno = rollno
        self.name = name
        self.pointer = pointer

    def __str__(self):
      # String concatenation
      # int and float values concatenated to str
      return "Name:"+self.name+" Rollno:"+str(self.rollno)+" CGPA:"+str(self.pointer)  

    def __del__(self):
      print("RIP",self.name)

# 10 Student entry
# for n till 1st blank 
slist=[]
rno=0
while True:
  
    rno+=1
    name=input("Enter name:")
    if name == '':
      break
    pointer=input("Enter pointer")
    s=Student(name,int(rno),float(pointer))
    slist.append(s)
    print("Student added to list\n Total students in list ",len(slist))
print("\nstudent list...")
for record in slist:
    print(record)
```
